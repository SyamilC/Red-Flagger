```python
import json
import joblib
import numpy as np

# 1. Custom JSON Encoder to catch any stray numpy types
class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NpEncoder, self).default(obj)

# 2. Load the saved pipeline
pipeline = joblib.load("../models/nlp_tfidf_logreg_model.pkl")
tfidf = pipeline.named_steps['tfidf']
classifier = pipeline.named_steps['classifier']

# 3. Extract and explicitly cast parameters to native Python types
# Explicitly cast vocab values to standard Python int
vocab = {str(k): int(v) for k, v in tfidf.vocabulary_.items()}

# .tolist() converts the numpy array to a list of native Python floats
idf = tfidf.idf_.tolist()

# Extract coefficients and intercepts, ensuring native types
coefs = [est.coef_[0].tolist() for est in classifier.estimators_]
intercepts = [float(est.intercept_[0]) for est in classifier.estimators_]

labels = [
    "aggressive_tone", 
    "hookup_focus", 
    "negativity", 
    "sarcasm_cynicism", 
    "substance_risk", 
    "incomplete_profile"
]

model_data = {
    "vocab": vocab,
    "idf": idf,
    "coefs": coefs,
    "intercepts": intercepts,
    "labels": labels
}

# 4. Save using the custom encoder
with open("../models/nlp_model.json", "w") as f:
    json.dump(model_data, f, cls=NpEncoder)

print("NLP model exported successfully to nlp_model.json")
```

```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Load data
df = pd.read_csv("../data/okcupid_cleaned_redflags.csv")
tabular_features = ["age", "status", "sex", "orientation", "body_type", "diet", "drinks", "drugs", "education", "ethnicity", "height", "income", "job", "offspring", "pets", "religion", "sign", "smokes"]
label_columns = ["aggressive_tone", "hookup_focus", "negativity", "sarcasm_cynicism", "substance_risk", "incomplete_profile"]

X = df[tabular_features]
y = df[label_columns].values

# Preprocess
numeric_features = ["age", "height", "income"]
categorical_features = [c for c in tabular_features if c not in numeric_features]

preprocessor = ColumnTransformer([
    ("num", StandardScaler(), numeric_features),
    ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
])

X_processed = preprocessor.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)

# Build a simple TF.keras Multi-Layer Perceptron (Drop-in replacement for RF in TF.js)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(len(label_columns), activation='sigmoid') # Multi-label output
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1, verbose=1)

# Save in TensorFlow.js Layers format
import tensorflowjs as tfjs
tfjs.converters.save_keras_model(model, "../models/tabular_model_tfjs")
print("Tabular model exported to tabular_model_tfjs/")
```
